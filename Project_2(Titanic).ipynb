```python
df['Age'] =df['Age'].fillna(df['Age'].median(),inplace=False)
df['Age']
```

```python
df.drop('Cabin',axis = 1,inplace = True)
print(df.isnull().sum())
```

```python
df.dropna(subset=['Embarked'],inplace = True)
print(df.isnull().sum())
```

```python
sns.countplot(x='Survived',palette=['Red','green'],data=df,edgecolor='black',width=.3)
plt.show()
```

```python
sns.countplot(x='Survived' ,hue= 'Sex',palette=['red','green'],data=df, width=.7,edgecolor='black')
```

```python
df.groupby(['Sex']).Survived.sum()
```

```python
sns.countplot(x='SibSp', data=df, palette='Set1', width=.9,edgecolor='black' )
```

```python
plt.figure(figsize=(5,5))
flierprops = dict(marker='D', markerfacecolor='yellow', markersize=5, linestyle='none')
sns.boxplot(x='Pclass', y='Age', palette='Set1', data=df, flierprops=flierprops)
```

```python
sns.histplot(x='Age',bins=20 ,data=df , color='red', edgecolor='black')
```

```python
numeric_df = df.select_dtypes(include=[float, int])

# Compute the correlation matrix
correlation = numeric_df.corr()

# Plot the heatmap
plt.figure(figsize=(15, 8))
sns.heatmap(correlation, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```
