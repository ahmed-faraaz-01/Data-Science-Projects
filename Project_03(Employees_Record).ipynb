```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime
import warnings
warnings.filterwarnings("ignore")
```

```python
df=pd.read_csv("C:/Users/STARHIVE/Downloads/employees.csv")
pd.set_option('display.max_rows', None)
df
```

```python
print('The no of rows and columns are: ',df.shape)
#df=df.shape[0]
#print('No of rows:',df)
```

```python
df.duplicated()
```

```python
df=df.drop_duplicates()
df.duplicated()
```

```python
df.duplicated().sum()
```

```python
df.isnull().sum()
```

```python
median=df['Age'].median()
df['Age']= df['Age'].fillna(median)
df.isnull().sum()

df['Department']=df['Department'].fillna('IT')
df.isnull().sum()
```

```python
df['Gender']=df['Gender'].replace({'Male': 'M', 'Female': 'F'})
df['Gender']
```

```python
df['Full Name']=df['First Name']+' '+df['Last Name']
df
columns=df.columns.tolist()
lastnameInd=columns.index('Last Name')
lastnameInd
columns.insert(lastnameInd +1 , columns.pop(columns.index('Full Name')))
df=df[columns]
df
```

```python
# Make columns as yearAtCompany
hiredate=pd.to_datetime(df['Hire_Date']).dt.year

curr_yr= datetime.now().year

df['YearsAtCompany']=curr_yr-hiredate
df.head(1)
```

```python
#bar Plot 
plt.figure(figsize=(10,6))
sns.countplot(x='Department', data=df ,  palette='Set1' )
plt.grid(True)
```

```python
department_count=df['Department'].value_counts()
plt.figure(figsize=(10,6))
color=['darkblue','darkgreen','yellow','purple']
department_count.plot(kind='bar',color=color)
plt.title('Emoloyee Count By department',fontweight='bold')
plt.xlabel('Department',fontweight='bold')
plt.ylabel('Emoloyee Count',fontweight='bold')
plt.xticks(rotation=0)
plt.show()
```
